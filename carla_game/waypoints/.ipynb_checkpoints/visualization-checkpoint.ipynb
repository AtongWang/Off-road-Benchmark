```python
from reward import Teacher
import matplotlib.pyplot as plt
import numpy as np
import random
import os
print(os.getcwd())
```

```python
city_name = "Offroad_1"
teacher = Teacher(city_name, ENV_INIT_VALUES[city_name][0])
```

```python
noise = np.random.normal(0, 10, size=(2))

idx = random.randint(0, len(teacher.wps))
location = teacher.wps[idx] + noise
```

```python
prev_points, next_points = teacher.get_nearest_waypoints(location, 50)
```

```python
points_group = {
    'waypoints': [teacher.wps[:, 0], teacher.wps[:, 1], 10],
    'previous waypoints': [prev_points[:, 0], prev_points[:, 1], 15],
    'next waypoints': [next_points[:, 0], next_points[:, 1], 15],
    'location': [[location[0]], [location[1]], 20]
}
```

```python
fig, ax = plt.subplots(figsize = (20, 20))
for key in points_group.keys():
    ax.plot(
        points_group[key][0], 
        points_group[key][1], 
        marker='o', 
        linestyle='',
        markersize=points_group[key][2],
        label=key
    )

'''
minv = int(np.min(teacher.wps))
maxv = int(np.max(teacher.wps))
width = maxv - minv
x = np.linspace(minv,maxv,width)
y = linear(x)
filters = (y < maxv) & (y > minv)
y = y[filters]
x = x[filters]
ax.plot(x, y, label='deadline')
'''

ax.legend(fontsize=20, loc='upper left') # legend position
plt.show()
```
